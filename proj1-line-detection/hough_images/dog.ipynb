```python
import matplotlib.pyplot as plt
import imageio
import numpy as np
from scipy.ndimage.filters import gaussian_filter
```

```python
blocks = imageio.imread('blocks.png')
plt.imshow(blocks, cmap='gray')
plt.show()
```

```python
blurred3 = gaussian_filter(blocks, sigma=7)
blurred7 = gaussian_filter(blocks, sigma=9)
diff = blurred3 - blurred7

plt.imshow(blurred3, cmap='gray')
plt.show()

plt.imshow(blurred7, cmap='gray')
plt.show()

plt.imshow(diff, cmap='gray')
plt.show()
```
